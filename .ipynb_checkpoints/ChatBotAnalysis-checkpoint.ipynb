### Training of the data
We shall be analysing the training data and also fine tuning the parameters how to train the data best 

We are going to begin by changing the dense factors first what is a dense factor.

This is the dense function we shall be using for training the data.
        
   y = $ \sum_{n=-\infty}^{+\infty} (w * x) + b $
   

We also use supervised learning for this dataset we can view the diagram below to see how it exactly works

![alt text](deep_learning_images.png)

###### observations
these are the contents about which we want to predict we shall use x for our observations
###### targets
these is what we are now predicting
##### model
this is the mathematical function that we shall apply to the observations in order to get 
the value of the targets
##### parameters
these are the weights for this situation we shall use w for the weights
##### predictions
the values of the targets that we get from the model.
#### loss function
the loss function tells us how far we are from the real values and try to reach an acceptable threshold






```python
train_data.tune_data(
        dense_factor_1=256,dense_factor_2=128,dense_factor_3=64,
        activation_1='relu',activation_2='relu',activation_3='relu',activation_4='softmax',
        dropout_factor_1=0.5,dropout_factor_2=0.5,
        lr=0.01,momentum=0.9,nesterov=True,loss='categorical_crossentropy',
        epochs=200,batch_size=5,verbose=1 
)
```
